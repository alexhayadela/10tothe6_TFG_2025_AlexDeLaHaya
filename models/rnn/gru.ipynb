```python
import torch
import torch.nn as nn
import torch.optim as optim

# Example: predicting next-day stock price from past 10 days
input_size = 1       # e.g., closing price only
hidden_size = 32
num_layers = 2
output_size = 1
seq_length = 10

# Dummy data (replace with real stock prices)
x_train = torch.randn(64, seq_length, input_size)
y_train = torch.randn(64, output_size)

# GRU Model
class GRUModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        out, _ = self.gru(x)           # GRU outputs (all timesteps)
        out = out[:, -1, :]            # Take last timestep
        out = self.fc(out)             # Fully connected to output
        return out

model = GRUModel()

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
for epoch in range(50):
    optimizer.zero_grad()
    outputs = model(x_train)
    loss = criterion(outputs, y_train)
    loss.backward()   # automatic backprop
    optimizer.step()
    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```
